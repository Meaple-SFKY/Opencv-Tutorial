```python
import cv2 as cv
import numpy as np
```

```python
kernel = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04]])

image = cv.imread('../../imgs/Chapter03/3.2/Snipaste.png')
dst = image
cv.filter2D(image, -1, kernel, dst)

cv.imwrite('../../imgs/Chapter03/3.5/filter_blur.png', dst)
```

```python
kernel_asymmetry = np.array([[-2, -1, 0],
                             [-1,  1, 1],
                             [ 0,  1, 2]])

cv.filter2D(image, -1, kernel_asymmetry, dst)
cv.imwrite('../../imgs/Chapter03/3.5/filter_blur_.png', dst)
```
